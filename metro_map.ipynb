```python
import folium
metro = pd.read_csv('https://raw.githubusercontent.com/gr-oll/SLO_LA_Olympics/main/datasets/Metro.csv')
# Create a map centered around Los Angeles
m = folium.Map(location=[34.0522, -118.2437], zoom_start=12)
# Add markers for each metro station
for _, row in metro.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"ID: {row['stop_id']}<br>Stop: {row['stop_name']}",
        icon=folium.Icon(color='lightblue')
    ).add_to(m)
# Display the map
m
```
